```python
import argparse
import numpy as np
import gudhi
import matplotlib.pyplot as plot
```

```python
sG3 = [
    [0., 315., 1695., 2020., 1391., 616., 605., 1521., 2509., 2488.],
    [315., 0., 1615., 2027., 1425., 692., 679., 1569., 2521., 2492.],
    [1695., 1615., 0., 1092., 2473., 2044., 2030., 2520., 1747., 1727.],
    [2020., 2027., 1092., 0., 1673., 2161., 2150., 1684., 1535., 1526.],
    [1391., 1425., 2473., 1673., 0., 1591., 1573., 1166., 2399., 2381.],
    [616., 692., 2044., 2161., 1591., 0., 76., 1474., 2463., 2443.],
    [605., 679., 2030., 2150., 1573., 76., 0., 1457., 2454., 2430.],
    [1521., 1569., 2520., 1684., 1166., 1474., 1457., 0., 2208., 2192.],
    [2509., 2521., 1747., 1535., 2399., 2463., 2454., 2208., 0., 80.],
    [2488., 2492., 1727., 1526., 2381., 2443., 2430., 2192., 80., 0.]
]



# Convertir la lista a una matriz NumPy
sG3 = np.array(sG3)
```

```python
max_edge_length = 4000
max_dimension = 5
band = 0.0

print("#####################################################################")
print("RipsComplex creation from defined distance matrix G3")
   
rips_complex = gudhi.RipsComplex(
   distance_matrix=sG3, max_edge_length=max_edge_length  
)
```

```python
simplex_tree = rips_complex.create_simplex_tree(max_dimension=max_dimension)
   
message = "Number of simplices=" + repr(simplex_tree.num_simplices()) 
print(message)
  
diag = simplex_tree.persistence()

print("betti_numbers()=")
print(simplex_tree.betti_numbers())
```

```python
gudhi.plot_persistence_diagram(diag, band=band)
plot.show()
```

```python
gudhi.plot_persistence_barcode(diag)
plot.show()
```
